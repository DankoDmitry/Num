```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000

l = np.zeros(N, dtype=np.cdouble)
A = np.random.normal(0, np.sqrt(1/N), (N,N))
l = np.linalg.eig(A)[0]
    
t = np.linspace(-1, 1, N)
plt.scatter(l.real, l.imag)
plt.scatter(t, np.sqrt(1 - t**2), color='red')
plt.scatter(t, -np.sqrt(1 - t**2), color='red')
plt.grid()
plt.show()

print(max(abs(l)))
```

```python

N = 1000

l = np.zeros((N,2), dtype=np.cdouble)

for i in  range(N):
    A = np.random.normal(0, 1, (2,2))
    B = (A + A.transpose())/2
    l[i] = np.linalg.eig(B)[0]
    
print(np.shape(l))
print(l[0:10])
plt.scatter(l[:,0].real, l[:,0].imag)
plt.grid()
plt.show()
```
